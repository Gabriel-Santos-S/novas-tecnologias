# Classes

```python
class Veiculo:
    def __init__(self, marca, modelo, preco, ano):
        self.marca = marca
        self.modelo = modelo
        self.preco = preco
        self.ano = ano

    def imprimir(self):
        print(f"Marca {self.marca}")
        print(f"Modelo {self.modelo}")
        print(f"Preco {self.preco}")
        print(f"Ano {self.ano}")




```
