## Dicionario

```python
# Dicionario
carro1 = {
    "modelo" : "Monza GLS", 
    "ano": 1994, 
    "preco": 15000,
}

carro2 = {
    "modelo": "Kild", 
    "ano": 2026, 
    "preco": 6000,
}

# Imprimi dicionario inteiro
print (carro1)

# Imprimi chave especifica
print (carro1["modelo"])
```

## Função

```python

def valor_parrcela(carro_preco, taxa_financiamento, quantidade_parce):
    return (carro_preco + (carro_preco * taxa_financiamento)) / quantidade_parce

def texto(texto):
    return print(texto)

valor_parrcela1 = valor_parrcela(carro1['preco'], 10.15, 48)
valor_parrcela2 = valor_parrcela(carro1['preco'], 20, 72)

texto(valor_parrcela1)
texto(valor_parrcela2)
```
