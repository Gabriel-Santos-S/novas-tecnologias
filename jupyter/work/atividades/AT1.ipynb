# Exercício 1 - Cadastro e análise simples de veículos

Uma concessionária deseja registrar alguns veículos disponíveis para venda.
Crie um programa em Python que crie uma lista de veículos.

Cada veículo deve ser representado por um dicionário contendo:
- marca
- modelo
- preço
- ano

Exemplo de estrutura esperada:

```
veiculo = {
    "marca": "Toyota",
    "modelo": "Corolla",
    "preco": 120000,
    "ano": 2022
}
```

O programa deve:

- Adicionar pelo *menos 5 veículos* à lista.
- Exibir todos os veículos cadastrados.
- Calcular o *preço médio dos veículos*.
- Mostrar quais *veículos custam acima da média*.

### Saída esperada (exemplo)
```
Preço médio dos veículos: 85000
Veículos acima da média:
Honda Civic - 110000
Toyota Corolla - 120000
```


````python
# Solução do exercício 1

from typing import List

# Classe
class Veiculo:
    marca: str
    modelo: str
    preco: float
    ano: int

    def __str__(this):
        return f"""================
Marca => {this.marca}
Modelo => {this.modelo}
Ano => {this.ano}
R$: {this.preco:,.2f}
================\n"""
        
#Funções

def media(valor: float, totalItens: int) -> float:
    result: float = valor / totalItens
    return result

def carrosAcimaMedia(carrosList: List[Veiculo], media: float):
    carrosAcimaMedia: List[Veiculo] = []

    for carro in carrosList:
        if carro.preco > media:
            carrosAcimaMedia.append(carro)  
            
    print(f"""XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX
Carros Assima da média {media:,.2f}
XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX""")
        
    for carro in carrosAcimaMedia:
            print(carro)


# Main
quantidadeRegistros: int = 5
somaCarros: float = 0.00
carros: List[Veiculo] = []

for i in range(quantidadeRegistros):
    carro = Veiculo()
    
    print(f"\n\nVamos cadastrar o carro {1 + i}:")
    carro.marca = str(input("Digite a marca do carro: "))
    carro.modelo = str(input("Digite o modelo do carro: "))
    carro.preco = float(input("Digite o preço do carro: "))
    carro.ano = int(input("Digite o ano do carro: "))
    
    carros.append(carro)

for carro in carros:
    somaCarros += carro.preco
    print(carro)

media: float = media(somaCarros, quantidadeRegistros)

carrosAcimaMedia(carros, media)
    
````

# Exercício 2 - Sistema simples de financiamento

Uma concessionária oferece financiamento de veículos.

Crie uma função chamada:

```
calcular_financiamento(preco, entrada, parcelas)
```

A função deve calcular o valor financiado.

```
valor_financiado = preco - entrada
```

Considerar uma taxa de juros simples de 1.5% ao mês.

Fórmula simplificada:
```
valor_total = valor_financiado * (1 + taxa * parcelas)
parcela = valor_total / parcelas
```
A função deve retornar o valor da parcela.


### Crie um programa que:

- Defina o preço de um veículo
- Pergunte ao usuário: valor de entrada e número de parcelas
- Use a função para calcular o financiamento
- Exiba o valor da parcela
- Impedir parcelas maiores que 72 meses
- Mostrar uma mensagem caso isso ocorra (Número máximo de parcelas é 72.)

### Exemplo de saída

```
Preço do veículo: 80000
Entrada: 20000
Parcelas: 48
Valor da parcela: 1785.50
```


````python
# Solução do exercício 2

# Imports
import time


# Funções
def calcular_financiamento(preco: float, entrada: float, parcelas: int) -> float:
    valor_financiado = preco - entrada

    valor_total: float = valor_financiado * (1 + 0.015 * parcelas)
    valor_parcela: float = valor_total / parcelas

    return valor_parcela
    

# Main
import time


preco_veiculo: float = 45000.00

print(f"Valor do carro {preco_veiculo}\n")

user_valor_entrada = float(input("Digite o valor da entrada: "))

num_tentativas = 0

while True:

    if num_tentativas >= 5:
        print("Você já usou todas as tentativas para preencher os dados\nTente mais tarde")
        break
    
    user_num_parcelas = int(input("Digite o numero de parcelas: "))

    if user_num_parcelas <= 72:
        print("Calculando . . .")
        time.sleep(2)

        valor_parcela = calcular_financiamento(preco_veiculo, user_valor_entrada, user_num_parcelas)
        
        print(f"""Preço do veículo: R$ {preco_veiculo:,.2f}
Entrada: R$ {user_valor_entrada:,.2f}
Parcelas: {user_num_parcelas}
Valor da parcela: R$ {valor_parcela:,.2f} """)
        break

    else:
        print("Pode no máximo 72 parcelas")
        num_tentativas += 1
````
