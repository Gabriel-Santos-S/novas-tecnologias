# Exercício 1 - Cadastro e análise simples de veículos

Uma concessionária deseja registrar alguns veículos disponíveis para venda.
Crie um programa em Python que crie uma lista de veículos.

Cada veículo deve ser representado por um dicionário contendo:
- marca
- modelo
- preço
- ano

Exemplo de estrutura esperada:

```
veiculo = {
    "marca": "Toyota",
    "modelo": "Corolla",
    "preco": 120000,
    "ano": 2022
}
```

O programa deve:

- Adicionar pelo *menos 5 veículos* à lista.
- Exibir todos os veículos cadastrados.
- Calcular o *preço médio dos veículos*.
- Mostrar quais *veículos custam acima da média*.

### Saída esperada (exemplo)
```
Preço médio dos veículos: 85000
Veículos acima da média:
Honda Civic - 110000
Toyota Corolla - 120000
```


````python
from typing import List

# Classe
class Veiculo:
    marca: str
    modelo: str
    preco: float
    ano: int

    def __str__(this):
        return f"""================
Marca => {this.marca}
Modelo => {this.modelo}
Ano => {this.ano}
R$: {this.preco:,.2f}
================\n"""
        
#Funções

def media(valor: float, totalItens: int) -> float:
    result: float = valor / totalItens
    return result

def carrosAcimaMedia(carrosList: List[Veiculo], media: float):
    carrosAcimaMedia: List[Veiculo] = []

    for carro in carrosList:
        if carro.preco > media:
            carrosAcimaMedia.append(carro)  
            
    print(f"""XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX
Carros Assima da média {media:,.2f}
XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX""")
        
    for carro in carrosAcimaMedia:
            print(carro)


# Main
quantidadeRegistros: int = 5
somaCarros: float = 0.00
carros: List[Veiculo] = []

for i in range(quantidadeRegistros):
    carro = Veiculo()
    
    print(f"\n\nVamos cadastrar o carro {1 + i}:")
    carro.marca = str(input("Digite a marca do carro: "))
    carro.modelo = str(input("Digite o modelo do carro: "))
    carro.preco = float(input("Digite o preço do carro: "))
    carro.ano = int(input("Digite o ano do carro: "))
    
    carros.append(carro)

for carro in carros:
    somaCarros += carro.preco
    print(carro)

media: float = media(somaCarros, quantidadeRegistros)

carrosAcimaMedia(carros, media)
    
````
